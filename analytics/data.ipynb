```python
import csv
import numpy as np

# Abra o arquivo CSV para leitura
with open('data.csv', 'r') as csvfile:
    # Crie um leitor CSV
    csvreader = csv.reader(csvfile)

    # Initialize lists to store values
    q1_values = []
    q2_values = []

    # Initialize variables for max and min values
    max_q1 = float('-inf')
    min_q1 = float('inf')

    max_q2 = float('-inf')
    min_q2 = float('inf')

    # Itere pelas linhas do arquivo CSV
    for row in csvreader:
        if len(row) == 2:
            q1_parts = row[0].split('=')  # Divide a primeira coluna em partes
            q2_parts = row[1].split('=')  # Divide a segunda coluna em partes

            if len(q1_parts) == 2:
                q1_value = float(q1_parts[1])
                q1_values.append(q1_value)
                max_q1 = max(max_q1, q1_value)
                min_q1 = min(min_q1, q1_value)

            if len(q2_parts) == 2:
                q2_value = float(q2_parts[1])
                q2_values.append(q2_value)
                max_q2 = max(max_q2, q2_value)
                min_q2 = min(min_q2, q2_value)

    # Convert lists to NumPy arrays
    q1_values = np.array(q1_values)
    q2_values = np.array(q2_values)

    # Calculate mean, std, max, and min
    mean_q1 = np.mean(q1_values)
    std_q1 = np.std(q1_values)

    mean_q2 = np.mean(q2_values)
    std_q2 = np.std(q2_values)

    print(f"Q1: Mean={mean_q1:.2f}, Std={std_q1:.2f}, Max={max_q1:.2f}, Min={min_q1:.2f}")
    print(f"Q2: Mean={mean_q2:.2f}, Std={std_q2:.2f}, Max={max_q2:.2f}, Min={min_q2:.2f}")
```
