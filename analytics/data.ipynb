```python
import csv

# Abra o arquivo CSV para leitura
with open('data.csv', 'r') as csvfile:
    # Crie um leitor CSV
    csvreader = csv.reader(csvfile)

    # Inicialize as variáveis para armazenar os valores máximos e mínimos
    max_q1 = float('-inf')  # Inicialize com o menor valor possível
    min_q1 = float('inf')   # Inicialize com o maior valor possível
    max_q2 = float('-inf')
    min_q2 = float('inf')

    # Itere pelas linhas do arquivo CSV
    for row in csvreader:
        if len(row) == 2:
            q1_parts = row[0].split('=')  # Divide a primeira coluna em partes
            q2_parts = row[1].split('=')  # Divide a segunda coluna em partes
            
            if len(q1_parts) == 2:
                q1_value = float(q1_parts[1])
                max_q1 = max(max_q1, q1_value)
                min_q1 = min(min_q1, q1_value)
            
            if len(q2_parts) == 2:
                q2_value = float(q2_parts[1])
                max_q2 = max(max_q2, q2_value)
                min_q2 = min(min_q2, q2_value)

# Agora você tem os valores máximos e mínimos de q1 e q2
print(f"Valor máximo de q1: {max_q1}")
print(f"Valor mínimo de q1: {min_q1}")
print(f"Valor máximo de q2: {max_q2}")
print(f"Valor mínimo de q2: {min_q2}")
```
